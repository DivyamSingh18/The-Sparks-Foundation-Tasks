```python
plt.figure(figsize=(8,8))
corr = df.corr()
sns.heatmap(corr,annot=True,cmap='Greens')
```

### We can observe that the 'Sales' is highly correlated with Profit , So in order to increase the profit, the sales must be increased
### Discount has Negative correlation with Profit

```python
df['Ship Mode'].value_counts()
```

```python
sns.countplot(x=df['Ship Mode'])
```

### We can observe that the most common ship mode is 'Standard Class' and the least common ship mode is 'Same Day' . <br><br> The standard class may have been less expensive. We can also get people into buying into the Same day options by offering discounts and coupens. 

```python
sns.countplot(x = 'Segment',data = df, )
```

### The Most common segment is Consumer and the least common segment is Home Office

```python
df['Sub-Category'].value_counts()
```

### We should focus on the quality of products that are bought less.

```python
plt.figure(figsize=(15,12))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

### binders, Paper, Furnishings and phones sub categories are larger than the sub categories like tables, envelops, supplies etc.

```python
df['State'].value_counts()
```

### Our company can offer special kind of products for specific states to increase the sales from these states

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=df,palette='rocket_r',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

### Most of the rows contain California ie highest buyers are from California

```python
plt.figure(figsize=(10,10))
explode = [0.1,.1,.1,.1]
df['Region'].value_counts().plot.pie(autopct = '%1.1f%%', explode = explode)
plt.show()
```

### The least buyers are from South in the dataframe

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','orange'],figsize=(15,8))
plt.title('Profit/loss vs Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

```python
p = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
```

### The most Profit comes from California and New York , The most Loss comes from Texas, Pennsylvania and Ohio

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','purple'],alpha=0.9,figsize=(12,9))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

### The profit of Furniture is very less as compared to the Sales

```python
p
```

```python
p[:].plot.bar(color=['blue','Green'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

### Phones sub category is having the highest sales but the highest profit is from the Copiers and Tables is at a huge loss. The Tables sub category should be targeted through advertising, should be made less costly 

```python
plt.figure(figsize = (5,5))
textprops = {"fontsize":10}
plt.title('Category')
plt.pie(df['Category'].value_counts(), labels=df['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.lineplot(df['Discount'], df['Profit'], data=df)
```

### No correlation of Discount with Profit 

```python
df.plot(kind = "scatter",figsize = (13,8), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "viridis")
plt.ylabel('Total Profits',fontsize = 16)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 16)

plt.show()
```

### The Profit increases with the Sales but the Profit is decreasing with the increasing discount
